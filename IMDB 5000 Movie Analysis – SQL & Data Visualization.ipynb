# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Best-Rated Directors (Min. 5 Movies)

```python
# Load data
df_directors = pd.read_csv("Best-Rated Directors.csv")

df_directors.dropna(inplace=True)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(
    df_directors['director_name'], 
    df_directors['avg_rating'],
    color='#3498db'
)
plt.xlabel('Average IMDB Rating')
plt.title('Top 10 Directors by Average Movie Rating')
plt.gca().invert_yaxis()  # Highest rating at top
plt.grid(axis='x', linestyle='--')
plt.show()
```

# Genre Profitability Analysis

```python
# Load data
df_genres = pd.read_csv("Genere based Profit.csv")

print(df_genres.head())

plt.figure(figsize=(12, 6))
df_genres['primary_genre'] = df_genres['genres'].str.split('|').str[0]  # Simplify
df_genres.sort_values('avg_profit', ascending=False, inplace=True)
sns.barplot(
    x='primary_genre', 
    y='avg_profit', 
    data=df_genres, 
    palette='viridis'
)
plt.xticks(rotation=45)
plt.ylabel('Avg Profit (USD)')
plt.title('Profitability by Primary Genre')
plt.show()
```

# Top gross movies

```python
# Load data
df_top_movies = pd.read_csv("Top gross movies.csv")

# Plot
plt.figure(figsize=(12, 6))
plt.stem(
    df_top_movies['movie_title'],
    df_top_movies['profit'] / 1e6,  # Scale to millions
    linefmt='grey',
    markerfmt='D',
    basefmt=' '
)
plt.xticks(rotation=90)
plt.ylabel('Profit (Millions USD)')
plt.title('Top 10 Highest-Grossing Movies')
plt.tight_layout()
plt.show()
```
